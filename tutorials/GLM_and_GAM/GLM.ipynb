# Why GLMs were introduced?
<hr style = "border:2px solid black" ></hr>

<div class="alert alert-info">
<font color=black>
    
- The linear regression model assumes that the outcome given the input features follows a Gaussian distribution. - This assumption excludes many cases: The outcome can also be a category (cancer vs. healthy), a count (number of children), the time to the occurrence of an event (time to failure of a machine) or a very skewed outcome with a few very high values (household income). 
- The linear regression model can be extended to model all these types of outcomes. This extension is called Generalized Linear Models or GLMs for short. 

</font>
</div>

# Introduction to GLMs
<hr style = "border:2px solid black" ></hr>

<div class="alert alert-info">
<font color=black>

- The generalised linear model (GLM) is a flexible generalisation of ordinary linear regression that allows for response variables that have error distribution models other than a normal distribution. 
- In a generalised linear model (GLM), each outcome Y of the dependent variables is assumed to be generated from a particular distribution in an exponential family, a large class of probability distributions that includes the normal, binomial, Poisson and gamma distributions, among others.
- **In brief**: with Generalized Linear Models, one uses a common training technique for a diverse set of regression models.
- GLMs allow the modeller to express the relationship between the regression variables X and the response variable (a.k.a. dependent variable) y, in a linear and additive way even though the underlying relationships may be neither linear nor additive.
                                                                                                                   
</font>
</div>

# GAMs svs. GLMs
<hr style = "border:2px solid black" ></hr>

<div class="alert alert-info">
<font color=black>

- GLM  (Generalised Linear Model) & GAM (Generalised Additive Mode).
- GAMs are an extension of GLMs and are more flexible than GLMs.
- The main difference is that GLMs models assume a fixed linear or some other parametric form of the relationship between the dependent variable and the covariates, GAM do not assume a priori any specific form of this relationship, and can be used to reveal and estimate non-linear effects of the covariate on the dependent variable. 
- GAMs have the advantage of being completely automatic, i.e. no "detective" work is needed on the part of the statistician.

</font>
</div>

# Poisson regression
<hr style = "border:2px solid black" ></hr>

<div class="alert alert-info">
<font color=black>

- This regressor is well suited for predicting counts. How many rain droplets a square foot gets in a year? How many clicks a link get in a day? How many bids an item gets in an auction? As you can see, all of these are counts of events happening within a specific range, time, area, etc. Of course, some other use-cases other than counts can also call for the Poisson Regression. 
- But the **key** is that we are predicting non-negative integers.

</font>
</div>

```python
def exp(x): return np.exp(x)[0]


x = np.array([[i] for i in np.random.choice(range(1, 10), 100)])


y = np.array([exp(i) + np.random.normal(0, 0.1*exp(i), 1)[0] for i in x])
y[y < 0] = 0
```

```python
from sklearn.linear_model import PoissonRegressor
pr = PoissonRegressor(alpha=0, fit_intercept=False)
y_pred_pr = pr.fit(x, y).predict(x)
```

```python
from matplotlib.pyplot import plot as plt
```

```python
plt(x,y, "o")
plt(x,y_pred_pr, "ro")
```

# Conclusions
<hr style = "border:2px solid black" ></hr>

<div class="alert alert-danger">
<font color=black>

- The GLMs extent the traditional ordinary least squares linear regression by adding a link function and assuming different distributions for the targets, as long as these distributions belong the exponential family of distributions.
- Plugging a link function allows the model to constraint its targets between 0 and 1 (in the case of the Logistic regression), above 0 (in the case of Poisson regression), or any other constraints depending on the link used. There are more GLMs that the ones discussed here. For example, Gamma and Inverse Gaussian.
- Finally, linear models are **good at extrapolation**, unlike tree-based models for example. Yet, they **dont have the ability** to capture feature interactions or their non-linearity
    
</font>
</div>
